# Análise de Dados do Mercado de Ações

O seguinte collab está sendo construído apenas com a finalidade de estudo, nele quero explorar algumas análises estatíticas básicas e, também, bibliotecas como Pandas e Yfinance.

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# Defina o símbolo da ação e o intervalo de datas desejado

symbol = 'ITUB4.SA'  # Símbolo da Itaú.
start_date = '2020-01-01'
end_date = '2023-11-01'
```

```python
# Obtenha os dados históricos
data = yf.download(symbol, start=start_date, end=end_date)

# Visualize os primeiros registros dos dados
print(data.head())
```

```python
# Calcular estatísticas básicas
print("Média de Preço de Fechamento:", data['Close'].mean())
print("Desvio Padrão de Preço de Fechamento:", data['Close'].std())
print("Preço Mínimo de Fechamento:", data['Close'].min())
print("Preço Máximo de Fechamento:", data['Close'].max())
```

```python
# Criar um gráfico de linha para os preços de fechamento
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Close'], label='Preço de Fechamento', color='blue')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')
plt.title('Variação do Preço de Fechamento ao Longo do Tempo')
plt.legend()
plt.show()
```
